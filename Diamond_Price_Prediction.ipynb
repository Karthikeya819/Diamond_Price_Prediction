```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import Module
from Module import Linear_Regression
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df.drop(columns=['cut','color','clarity']).corr() # No need to consider depth,table in determination of price
```

```python
df = pd.get_dummies(df.drop(columns=['depth','table'])).astype(float)
df
```

# Normalization

```python
Norm_df = Module.Normalize(df,df.columns)
df_train, df_test = Module.train_test_split(Norm_df, test_size=0.2, random_state=42)
```

# Training Model

```python
X_train = df_train.drop(columns=['price'])
Y_train = df_train['price']
model = Linear_Regression(0.01,1000)
model.fit(X_train,Y_train)
```

# Prediction

```python
X_test = df_test.drop(columns=['price'])
Y_test = df_test['price']
result = model.predict(X_test)
R2 = Module.R2(result,Y_test)
print('Accuracy of Model is '+str(R2*100)+' %')
```
